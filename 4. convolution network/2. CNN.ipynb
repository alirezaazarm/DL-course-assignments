```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Flatten
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.initializers import lecun_normal
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import LearningRateScheduler
```

```python
# Load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
# Split the training dataset into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.15, random_state=42)
```

```python
x_train.shape
```

#first CNN model

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(100, 5, activation="relu",
                         padding="same", input_shape=[32, 32, 3]),
    keras.layers.AveragePooling2D(2),
    keras.layers.Conv2D(50, 3, activation="relu", padding="same"),
    keras.layers.AveragePooling2D(2),
    keras.layers.Conv2D(25, 2, activation="relu", padding="same"),
    keras.layers.MaxPooling2D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation="selu",  kernel_initializer=lecun_normal()),
    keras.layers.Dense(10, activation="softmax")
])

model.compile(loss="sparse_categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])

model1 = model.fit(x_train, y_train, epochs=50, validation_data=[x_val, y_val], batch_size=64)
```

# **One cycle time scheduling learning rate**

```python
# Define your learning rate schedule
def lr_schedule(epoch, lr):
    initial_lr = 0.001
    end_lr = 0.00001
    warmup_epochs = 5
    total_epochs = 50

    if epoch < warmup_epochs:
        lr = initial_lr + epoch * (initial_lr - end_lr) / warmup_epochs
    elif epoch >= warmup_epochs and epoch <= total_epochs:
        lr = 2*initial_lr - end_lr + (epoch - warmup_epochs) * 2 * (end_lr - initial_lr) / (total_epochs - warmup_epochs)
    else:
        lr = end_lr

    return lr
```

```python
lr_scheduler = LearningRateScheduler(lr_schedule)
```

```python
model_cycle = model.fit(x_train, y_train, epochs=50, validation_data=[x_val, y_val], batch_size=64, callbacks=[lr_scheduler])
```

#exponential time scheduling

```python
# Define the exponential decay function for learning rate scheduling
def exponential_decay(epoch, initial_lr, decay_rate, decay_steps):
    return initial_lr * decay_rate**(epoch // decay_steps)

initial_learning_rate = 0.001
decay_rate = 0.5
decay_steps = 5

# Create a learning rate scheduler callback
exp_lr_scheduler = LearningRateScheduler(lambda epoch: exponential_decay(epoch, initial_learning_rate, decay_rate, decay_steps))

model.fit(x_train, y_train, epochs=50, validation_data=[x_val, y_val], batch_size=64, callbacks=[exp_lr_scheduler])
```

#transformed model (Xception)

```python
import numpy as np

new_x = np.array(tf.image.resize(x_train[:3000,], [299, 299]))
```

```python
new_y = y_train[:3000,]
```

```python
new_y.shape
```

```python
xception = keras.applications.Xception()

xception_model = keras.applications.Xception(include_top=True)
output = keras.layers.Dense(10, activation="softmax")(xception_model.output)
model = keras.Model(inputs=xception_model.input, outputs=output)

for layer in xception_model.layers:
  layer.trainable = False

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(new_x, new_y, epochs=50, validation_split=0.15, batch_size=64)
```

# just adding global average pooling layer

```python
xception = keras.applications.Xception()

xception_model = keras.applications.Xception(include_top=False)
ga_layer = keras.layers.GlobalAveragePooling2D()(xception_model.output)
output = keras.layers.Dense(10, activation="softmax")(ga_layer)
model = keras.Model(inputs=xception_model.input, outputs=output)

for layer in xception_model.layers[:-1]:
  layer.trainable = False

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(new_x, new_y, epochs=50, validation_split=0.15, batch_size=64)
```

```python
from tensorflow.keras.applications import Xception

model = Xception(weights='imagenet', include_top=True)

# Set all layers except the last layer as untrainable
for layer in model.layers[:-1]:
    layer.trainable = False

last_layer = model.layers[-1]

# Create a new output layer with the desired number of classes
num_classes = 10  # Replace with the desired number of output classes
output = Dense(num_classes, activation='softmax')(last_layer.output)

# Create the modified model with the new output layer
modified_model = keras.Model(inputs=model.input, outputs=output)

# Compile the model and specify the desired optimizer and loss function
modified_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
modified_model.fit(new_x, new_y, epochs=50, validation_split=0.15, batch_size=64)
```
