```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def selu(x, alpha=1.67326, scale=1.0507):
    """Scaled Exponential Linear Unit (SELU) activation function."""
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

def relu(x):
    """Rectified Linear Unit (ReLU) activation function."""
    return np.maximum(0, x)

def sigmoid(x):
    """Sigmoid activation function."""
    return 1 / (1 + np.exp(-x))

def tanh(x):
    """Hyperbolic tangent (tanh) activation function."""
    return np.tanh(x)

# Define the derivatives of the activation functions
def selu_deviation(x, alpha=1.67326, scale=1.0507):
    """Derivative of the SELU activation function."""
    return scale * np.where(x > 0, 1, alpha * np.exp(x))

def relu_deviation(x):
    """Derivative of the ReLU activation function."""
    return np.where(x > 0, 1, 0)

def sigmoid_deviation(x):
    """Derivative of the sigmoid activation function."""
    return sigmoid(x) * (1 - sigmoid(x))

def tanh_deviation(x):
    """Derivative of the tanh activation function."""
    return 1 - np.square(np.tanh(x))

# Generate data points
x = np.linspace(-10, 10, 1000)
y_selu = selu(x)
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_selu_dev = selu_deviation(x)
y_relu_dev = relu_deviation(x)
y_sigmoid_dev = sigmoid_deviation(x)
y_tanh_dev = tanh_deviation(x)

# Increase plot size
plt.figure(figsize=(15, 10))


# Plot the activation functions
plt.plot(x, y_selu, label='SELU')
plt.plot(x, y_relu, label='ReLU')
plt.plot(x, y_sigmoid, label='sigmoid')
plt.plot(x, y_tanh, label='tanh')


# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

```python

# Increase plot size
plt.figure(figsize=(15, 10))

# Plot the derivatives of the activation functions
plt.plot(x, y_selu_dev, label='SELU derivation')
plt.plot(x, y_relu_dev, label='ReLU derivation')
plt.plot(x, y_sigmoid_dev, label='sigmoid derivation')
plt.plot(x, y_tanh_dev, label='tanh derivation')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
